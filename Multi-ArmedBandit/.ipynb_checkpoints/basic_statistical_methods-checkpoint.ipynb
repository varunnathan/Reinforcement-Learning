```python
import numpy as np
import matplotlib.pyplot as plt
```

## Epsilon-greedy

```python
class Bandit(object):
    
    def __init__(self, m):
        self.m = m
        self.N = 0
        self.mean = 0
        
    def pull(self):
        return np.random.randn() + self.m
    
    def update(self, x):
        self.N += 1
        self.mean = (1-1/self.N)*self.mean + (1/self.N)*x

def run_experiment(m1, m2, m3, eps, N):
    bandits = [Bandit(m1), Bandit(m2), Bandit(m3)]
    data = np.empty(N)
    
    for i in range(N):
        p = np.random.random()
        if p < eps:
            j = np.random.choice(3)
        else:
            j = np.argmax([b.mean for b in bandits])
        
        x = bandits[j].pull()
        bandits[j].update(x)
        data[i] = x
    
    cumulative_avg = np.cumsum(data)/(np.arange(N)+1)
    
    plt.plot(cumulative_avg)
    plt.plot(np.ones(N)*m1)
    plt.plot(np.ones(N)*m2)
    plt.plot(np.ones(N)*m3)
    plt.xscale('log')
    plt.show()
    
    for i, b in enumerate(bandits):
        print('Bandit %d: ' %(i+1), b.mean)
    
    return cumulative_avg
```

```python
c_1 = run_experiment(1.0, 2.0, 3.0, 0.1, 100000)
```

```python
c_05 = run_experiment(1.0, 2.0, 3.0, 0.05, 100000)
```

```python
c_01 = run_experiment(1.0, 2.0, 3.0, 0.01, 100000)
```

```python
c_2 = run_experiment(1.0, 2.0, 3.0, 0.2, 100000)
c_3 = run_experiment(1.0, 2.0, 3.0, 0.3, 100000)
```

```python
# log scale plot
plt.plot(c_3, label='eps = 0.3')
plt.plot(c_2, label='eps = 0.2')
plt.plot(c_1, label='eps = 0.1')
plt.plot(c_05, label='eps = 0.05')
plt.plot(c_01, label='eps = 0.01')
plt.legend()
plt.xscale('log')
plt.show()
```
